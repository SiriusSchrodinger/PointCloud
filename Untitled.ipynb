```python
import h5py
import torch
from pykeops.torch import LazyTensor
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.mixture import GaussianMixture
from matplotlib.patches import Ellipse
import numpy as np
from math import pi, cos, sin

f = h5py.File('./MNIST data/train_0.h5', 'r')
data = f['data']
label = f['label']
```

```python
selected_data = data[2]

selected = selected_data[:, :3]
#selected = selected_data
fig = plt.figure()
ax = Axes3D(fig)
xs = selected[:, 0]
ys = selected[:, 1]
zs = selected[:, 2]
#intensity = selected[:, 3]
ax.scatter(xs, ys, zs)
ax.view_init(elev=10., azim=-80)
plt.show()
print(label[2])
```

```python
gmm = GaussianMixture(n_components=15)
gmm.fit(selected)
means = gmm.means_
covariances = gmm.covariances_
```

```python
new_covariances = np.zeros(60).reshape((15, 2, 2));
for i in range(15):
    new_covariances[i][0][0] = covariances[i][0][0]
    new_covariances[i][0][1] = covariances[i][0][2]
    new_covariances[i][1][0] = covariances[i][2][0]
    new_covariances[i][1][1] = covariances[i][2][2]
mx = means[:, 0]
my = means[:, 1]
mz = means[:, 2]

fig, ax = plt.subplots()
ax.scatter(xs, zs, c = 'tab:blue', alpha = 0.3, label = "data")
ax.scatter(mx, mz, c = 'black', label = "center")
for i in range(15):
    u, s, vt = np.linalg.svd(new_covariances[i])
    t = np.linspace(0, 2*pi, 100)
    Ell = np.array([np.sqrt(s[0]) * np.cos(t) , np.sqrt(s[1]) * np.sin(t)])
    Ell_rot = np.zeros((2,Ell.shape[1]))
    for j in range(Ell.shape[1]):
        Ell_rot[:,j] = np.dot(u,Ell[:,j])

    ax.plot(mx[i] + Ell_rot[0,:] ,mz[i] + Ell_rot[1,:], c = 'grey')
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.xlabel("x", fontsize=20)
plt.ylabel("z", fontsize=20)
ax.legend(fontsize=15)
plt.show()
```
